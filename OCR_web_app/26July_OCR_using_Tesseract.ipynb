#### Note

1) OCR is used to read text from the images.<br>
2) <b>It is not guaranteed to generate the correct text every time</b><br>
3) The preferred image should bear black blackground and the text should be in white and the text font 
should be same thoroughout<br>
4) The text(read) using OCR is in English<br>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pytesseract
```

```python
pytesseract.pytesseract.tesseract_cmd = "C:\\Program Files\\Tesseract-OCR\\tesseract.exe"
```

```python
img1 = cv2.imread('ocr5.png')
plt.imshow(img1)
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
text_img1 = pytesseract.image_to_string(img1)
print(text_img1)
```
